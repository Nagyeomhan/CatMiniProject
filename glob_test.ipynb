```python
# 프레임 넘버를 뽑아내기 위한 정규화 함수 미리 설정
def getFrameNum(imgcol):
    p = '_(\d{1,5})_'
    if re.search(p, imgcol):
        result = re.findall(p, imgcol)
        return result[0]
```

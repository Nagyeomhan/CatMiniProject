```python
import pandas as pd
import json
import os
```

```python
path = 'C:/Users/admin/Desktop/PetProject/20201028_cat-arch-000156.mp4'
```

```python
jpgtitle = os.listdir(path)
print(jpgtitle)
```

```python
with open('20201028_cat-arch-000156.mp4.json', encoding='utf-8') as jsonfile:
    data = json.load(jsonfile)
df = pd.json_normalize(data['annotations'])
df = df.drop(columns=['frame_number', 'frame_url', 'timestamp'])
df
```

```python
df.columns
```

```python
# 반복문으로 프레임당 메모장 하나씩 저장하기 테스트 코드
# 오류메세지가 뜨지만 제대로 작동하는 것을 확인했다
for i in range(len(df)):
    rowdf = df.iloc[i, :]
    rowdf.reset_index(drop=True)
    rowdf.to_csv(f'{jpgtitle[i]}.txt')
```

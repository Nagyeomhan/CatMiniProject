```python
# 파일 및 폴더 관련
import os
import glob
import shutil
import natsort
import os.path

# 작동 관련
import pandas as pd
import numpy as np
import json
```

```python
img_path = 'C:/Users/admin/Desktop/CAT/RAW/ARMSTRETCH/'
```

```python
folder_list = natsort.natsorted( os.listdir(img_path) )
folder_list
```

```python
file_list = natsort.natsorted( [items for items in os.listdir(img_path + folder_list[8])] )
file_list
```

```python
# 예시처럼 이름을 맞춘 프레임 이름들을 담을 리스트
framename = []

for x in file_list:
    y = folder_list[8] + '/' + x
    framename.append(y)

framename
```

```python
df = pd.DataFrame()
df['image'] = framename
df
```

```python
label_path = 'C:/Users/admin/Desktop/CAT/LABEL/ARMSTRETCH/'
json_files = [files for files in os.listdir(label_path) if files.endswith('.json')]
json_files
```

```python
with open('C:/Users/admin/Desktop/CAT/LABEL/ARMSTRETCH/20201111_cat-armstretch-000779.mp4.json', encoding='utf-8') as json_files:
    data = json.load(json_files)
df2 = pd.json_normalize(data['annotations'])
df2 = df2.drop(columns=['frame_number', 'frame_url', 'timestamp', 'bounding_box.x','bounding_box.y','bounding_box.width','bounding_box.height'])
df2
```
